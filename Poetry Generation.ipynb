##### Bigram CFD

```python
def Bigram_freq(tokens):
    freq_n={}
    in_f={}
    for i in range(len(tokens)):
        count=0
        index=[]
        for j in range(len(tokens)):
            if(tokens[i]==tokens[j]):
                index.append(j)
                count=count+1
        temp={}
        for k in index:
            counts=0
            if(k==len(tokens)-1):
                continue
            next_0=tokens[k+1]
            for l in index:
                if(l==len(tokens)-1):
                    continue
                next_1=tokens[l+1]
            
                if(next_0==next_1):
                    counts=counts+1
            in_f.clear()
            in_f[next_0]={}
            in_f[next_0]=counts/len(index)
            tk_text=tokens[i]
            if tk_text in freq_n:
                freq_n[tk_text].update(in_f)
            else:
                freq_n[tk_text]={}
                freq_n[tk_text].update(in_f)
    return freq_n
```

##### Trigam CFD

```python
def trigram_freq(Bigrams):
    freq_bi={}
    in_f={}
    for i in range(len(Bigrams)):
        count=0
        index=[]
        for j in range(len(Bigrams)):
            if(Bigrams[i]==Bigrams[j]):
                index.append(j)
                count=count+1
### counts and Index's fined
        temp={}
        for k in index:
            counts=0
            if(k==len(Bigrams)-1):
                continue
### spliting the Bigrams Next Word
            tn_0=Bigrams[k+1].split(' ')
            next_0=tn_0[1]
### Comparing the next word with other words for its frequency.
            for l in index:
                if(l==len(Bigrams)-1):
                    continue
                tn_1=Bigrams[l+1].split(' ')
                next_1=tn_1[1]
                if(next_0==next_1):
                    counts=counts+1
### storing the frequency of each word againt the BiGram 
            in_f.clear()
            in_f[Bigrams[k+1]]={}
            in_f[Bigrams[k+1]]=counts/len(index)
            tk_text=Bigrams[i]
            if tk_text in freq_bi:
                freq_bi[tk_text].update(in_f)
            else:
                freq_bi[tk_text]={}
                freq_bi[tk_text].update(in_f)
    return freq_bi
```

```python

def Back_gram(tokens):
    freq_back={}
    in_f={}
    for i in range(len(tokens)):
        count=0
        index=[]
        for j in range(len(tokens)):
            if(tokens[i]==tokens[j]):
                index.append(j)
                count=count+1
        temp={}
        for k in index:
            counts=0
            if(k==len(tokens)-1):
                continue
            next_0=tokens[k-1]
            for l in index:
                if(l==len(tokens)-1):
                    continue
                next_1=tokens[l-1]
            
                if(next_0==next_1):
                    counts=counts+1
            in_f.clear()
            in_f[next_0]={}
            in_f[next_0]=counts/len(index)
            tk_text=tokens[i]
            if tk_text in freq_back:
                freq_back[tk_text].update(in_f)
            else:
                freq_back[tk_text]={}
                freq_back[tk_text].update(in_f)

    return freq_back
    
```
